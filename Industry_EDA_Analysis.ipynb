# Outlier Detection

```python
#Boxplot Analysis - for all numerical values 
```

```python
num_cols = ['employee_count','annual_revenue_million',
            'profit_margin_percent','customer_count',
            'market_rating']

for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

# Key Insights:
-Very few extreme values were observed.
-Most variables appeared uniformly distributed.
-No significant outliers were visible in the boxplots.

```python
#Interquartile Range (IQR) Method

for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower) | (data[col] > upper)]

    print(col, len(outliers))
```

# Key Insights
-Since the dataset values are generated within predefined ranges, very few or no observations fall outside the IQR boundaries.
-No substantial outlier problem was detected.

```python
# Z-Score Method
from scipy.stats import zscore

for col in num_cols:
    z_scores = np.abs(zscore(data[col]))
    outliers = data[z_scores > 3]

    print(col, len(outliers))
```

# Key insights
-Very few records exceeded the Z-score threshold.
-The numerical variables showed stable distributions with minimal extreme values.

# Conclusion
-Boxplots indicated no major outlier concerns.
-The IQR method detected few or no significant outliers.
-The Z-score method confirmed the absence of extreme observations.
-Overall, the dataset appears well-balanced and does not require extensive outlier treatment before further analysis.
